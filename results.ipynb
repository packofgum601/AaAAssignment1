```python
import matplotlib.pyplot as plt
import numpy as np

x = ['Array', 'Linked List', 'Trie']
y1 = [0.7655267858505249, 1.4749501943588257, 0.20348347187042237]
y2 = [2.3657443284988404, 2.9237890100479125, 0.4203815031051636]
y3 = [10.407648930549621, 5.9598699569702145, 0.8782468557357788]

colors = ['red', 'green', 'blue']
# Create a figure with multiple subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns of subplots

# Create the first subplot (leftmost)
axes[0].bar(x, y1, color=colors)
axes[0].set_title('Delete Method 50k')
axes[0].set_xlabel('Categories')
axes[0].set_ylabel('Time Taken')

# Create the second subplot (middle)
axes[1].bar(x, y2, color=colors)
axes[1].set_title('Delete Method 100k')
axes[1].set_xlabel('Categories')
axes[1].set_ylabel('Time Taken')

# Create the third subplot (rightmost)
axes[2].bar(x, y3, color=colors)
axes[2].set_title('Delete Method 200k')
axes[2].set_xlabel('Categories')
axes[2].set_ylabel('Time Taken')

# Adjust spacing between subplots (optional)
plt.tight_layout()

# Show the entire figure with all subplots
plt.show()
```
